```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gen_paths(S0, mu, sigma):
    dt=1.0/365.0
    paths = np.zeros(3650)
    paths[0] = S0
    for t in range(1, 3650):
        rand = np.random.standard_normal(1)
        paths[t] = paths[t - 1] * np.exp((mu- 0.5 * sigma ** 2) * dt +
                                         sigma * np.sqrt(dt) * rand)
    return paths, dt
```

```python
s, dt =gen_paths(1,0.06,0.20)
```

```python
plt.plot(s,label = 'Price')
plt.grid()
plt.legend()
plt.xlabel('Days')
plt.ylabel('Stock price')
plt.title("10 years share price evolution")
plt.show()
```

```python
logret=np.log(s[1:3650]/s[0:3650-1])
```

```python
plt.plot(logret,label = 'Log returns')
plt.grid()
plt.legend()
plt.xlabel('Days')
plt.ylabel('Daily log-returns')
plt.title("10 years Daily log returns")
plt.show()
```

```python
print('The annualized mean is %1.4f'%(np.mean(logret)/dt))
print('The annualized standard deviation is %1.4f'%(np.std(logret)/np.sqrt(dt)))
```
